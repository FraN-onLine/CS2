```python
# do not change this cell/section
# run this section to initialize your dataset

import pandas as pd
import matplotlib.pyplot as plt

# google resource initialization
url = 'https://drive.google.com/file/d/10Bl75VAt2U3XTFoc8hBJbu9tvfvh7BMa/view?usp=sharing'
file_id=url.split('/')[-2]
dwn_url='https://drive.google.com/uc?id=' + file_id

# read the dataset
dataset = pd.read_csv(dwn_url)

dataset.drop(['Country Code', 'Indicator Name','Indicator Code'], inplace = True, axis=1)
dataset.head()
```

## Data Analysis

For your first objective, you need to compare and visualize the life expectancy of those countries with the happiest (`Finland`, `Denmark`, `Norway`) and saddest (`South Sudan`, `Central African Republic`, `Afghanistan`) individuals.  Discuss your perception of the figure and whether emotions might affect life expectancy.

```python

#the 6 countries for analysis
countries = ['Finland', 'Denmark', 'Norway', 'South Sudan', 'Central African Republic', 'Afghanistan']
subset = dataset[dataset['Country Name'].isin(countries)]

#transpose so x and y axises swap 
subset = subset.set_index('Country Name').transpose()

#make year into columns
subset = subset.reset_index().rename(columns={'index': 'Year'})

#convert to numerical data to avoid issues
subset['Year'] = subset['Year'].astype(int)

# line plot is efficient for this scenario
subset.plot.line(x='Year', marker='', figsize=(14, 7))

plt.title('Life Expectancy and Mood')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.xticks(subset['Year'], rotation=45)
plt.legend(title='Country')
plt.tight_layout()
plt.show()

```

```python
# A Line graph is used for this activity because and i quote- "Line graphs are most commonly used to plot continuous data.
# Because the points are physically connected via the line, it implies
# a connection between the points that may not make sense for categorical data *(a set of data that is sorted or divided into different
# categories)*. Often, our continuous data is in some unit of time: days,
# months, quarters, or years." - Baculpo, 2025

# and after being able to visualize the data, you can visually see a correlation between life expectancy and happiness
# Happier Countries shows higher life exoectancy
# Sadder Countries shows lower life expectancies
```

Your next objective is to compare and visualize the life expectancy of those countries with the best (`Luxembourg`, `Ireland`, `Singapore`) and worst (`Burundi`, `Somalia`, `Mozambique`) economic status in the world.  Also, discuss your perception of the figure and whether the economic status of the country might affect life expectancy.

```python

#the 6 countries for analysis
countries = ['Luxembourg', 'Ireland', 'Singapore', 'Burundi', 'Somalia', 'Mozambique']
subset = dataset[dataset['Country Name'].isin(countries)]

#transpose so x and y axises swap 
subset = subset.set_index('Country Name').transpose()

#make year into columns
subset = subset.reset_index().rename(columns={'index': 'Year'})

#convert to numerical data to avoid issues
subset['Year'] = subset['Year'].astype(int)

# line plot is efficient for this scenario
subset.plot.line(x='Year', marker='', figsize=(14, 7))

plt.title('Life Expectancy and Economic Status')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.xticks(subset['Year'], rotation=45)
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```
